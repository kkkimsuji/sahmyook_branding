기숙사는 안정적인 주거환경에서 교육받기 위한 필수시설이다. 특히 거주지와 대학 간 거리가 멀어 현실적으로 통학이 불가능한 지방출신 학생들에게 절실하다. 대학은 적정 규모의 기숙사를 건립하고 저렴한 비용으로 기숙시설을 제공해 학생들의 편의를 보장해야 할 의무가 있다. 정부는 대학생 기숙사 수용인원을 5만명까지 확충할 것을 국정과제로 제시한 상태다.
