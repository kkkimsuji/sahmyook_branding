```python
#기본패키지 설치
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

plt.rc('font', family = 'NanumBarunGothic') #한글폰트 사용할 수 있도록
```

```python
df = pd.read_csv('한국장학재단_대학별 장학금 수혜 현황_20220831 (1).csv', encoding = 'cp949')

df.tail()
```

```python
raw_data = df.copy() 
```

```python
df.describe()
```

```python
# 교내장학금 소계(원) 높은 순으로 정렬
df_sort = df.sort_values(by='교내장학금 소계(원)', ascending = False)
df_sort.head()
```

```python
# 순서대로 lank
lank = []
for i in range (len(df['학교명'])):
    lank.append(i+1)
```

```python
# lank부여
df_sort['교내장학금순위'] = lank
df_sort.head()
```

삼육대학교 장학금 분석

```python
s_data = df_sort[ df_sort['학교명']== '삼육대학교']
s_data
```

```python
s_m = float(s_data['교내장학금 소계(원)']/7062)
s_m
```

가천대학교 장학금 분석

```python
g_data = df_sort[ df_sort['학교명']== '가천대학교']
g_data
```

```python
g_m = float(g_data['교내장학금 소계(원)']/26126)
g_m
```

한성대학교 장학금 분석

```python
h_data = df_sort[ df_sort['학교명']== '한성대학교']
h_data
```

```python
h_m = float(h_data['교내장학금 소계(원)']/9484)
h_m
```
